```python
# geocodes by neighborhood name
# NaN becomes string null
# clean this up, super ugly code ick
import geocoder
longitude = []
latitude = []
n_index = 0
for neighbourhood in df['neighbourhood']:
    if pd.isnull(neighbourhood) == False and type(neighbourhood) != float:
        place = "%s, %s" % (neighbourhood, 'Noumea')
        g = geocoder.google(place)
        g.latlng
        try: 
            lat = g.lat
            long = g.lng
            latitude.append(lat)
            longitude.append(long)
        except: 
            lat = 'null'
            long = 'null'
            latitude.append(lat)
            longitude.append(long)
    else: 
        lat = 'null'
        long = 'null'
        latitude.append(lat)
        longitude.append(long)
    print(df.index[n_index], neighbourhood, lat, long)
    n_index = n_index + 1
df['longitude'] = longitude
df['latitude'] = latitude
```

```python
# convert years to months

ans = re.compile(r"ans*", re.IGNORECASE)
digit = re.compile(r"\d+", re.IGNORECASE)

def multiple_replacer(*key_values):
    replace_dict = dict(key_values)
    replacement_function = lambda match: replace_dict[match.group(0)]
    pattern = re.compile("|".join([re.escape(k) for k, v in key_values]), re.M)
    return lambda string: pattern.sub(replacement_function, string)

def multiple_replace(string, *key_values):
    return multiple_replacer(*key_values)(string)

replacements = (u"un", u"1"), (u"deux", u"2"), (u"trois", u"3"), (u"quatre", u"4"), (u"cinq", u"5"), (u"six", u"6"), (u"sept", u"7"), (u"huit", u"8"), (u"neuf", u"9"), (u"dix", u"10"), (u"j'y suis née", u"toujours")
# print(multiple_replace(stringie, *replacements))

months_arr = []

# case
# if the string contains years, multiply months * 12, remove 'ans'
for answer in df['months_lived_in_noumea']:
    has_ans = re.findall(ans, str(answer))
    print("initial", answer, "has ans:", has_ans, len(has_ans))
    if len(has_ans) == 1 and type(answer) != float:
        number = re.findall(digit, str(answer))
        print("here's the number", number)
        if len(number) == 1:
            new_answer = str(answer).lower()
            print("new answer:", new_answer)
            new_answer = multiple_replace(new_answer, *replacements)
            new_number = int(number[0]) * 12
            print("new number:", new_number)
            months_arr.append(str(new_number))
        else:
            new_answer = str(answer).lower()
            print("in first else")
            new_answer = multiple_replace(str(answer), *replacements)
            months_arr.append(str(new_answer))
    else:
        new_answer = multiple_replace(str(answer), *replacements)
        months_arr.append(str(new_answer))
    print(months_arr)
df['months_lived_in_noumea'] = months_arr
print(months_arr)
print(len(months_arr))
```
